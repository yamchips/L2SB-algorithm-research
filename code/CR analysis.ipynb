# Huffman vs L2SB

```python
data = {
    "Dataset": ["F", "O", "S", "MIT1", "MIT2"],
    "DPCM + Huffman": [2.130, 1.778, 1.398, 2.819, 2.697],
    "L2SB": [1.590, 1.379, 1.187, 1.636, 1.611]
}
df_compare = pd.DataFrame(data)
```

```python
# Plotting the data
fig, ax = plt.subplots()

# Define the bar width
bar_width = 0.35

# Define the positions of the bars
index = range(len(df_compare["Dataset"]))

# Create bars for DPCM + Huffman
bar1 = ax.bar(index, df_compare["DPCM + Huffman"], bar_width, label="DPCM + Huffman")

# Create bars for L2SB
bar2 = ax.bar([i + bar_width for i in index], df_compare["L2SB"], bar_width, label="L2SB")

# Add labels and title
ax.set_xlabel("Dataset")
ax.set_ylabel("Compression Ratio")
ax.set_xticks([i + bar_width / 2 for i in index])
ax.set_xticklabels(df_compare["Dataset"], rotation=45, ha="right")
ax.legend()
```
